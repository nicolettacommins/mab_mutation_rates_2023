# Plot fraction of samples with read depth > 20x across the reference (which regions tend to have good v. poor coverage?)

### MAB

```python
with open('depth_plots/mabdepth.data', 'rb') as filehandle:
    mabdepth=pickle.load(filehandle)
```

```python
x=list(range(1,len(mabdepth)+1))
```

```python
fig=plt.figure(figsize=(12,8))

mabfrac=[i/904 for i in mabdepth]

plt.plot(x, mabfrac, linewidth=0.05, color='blue')
plt.ylabel('Fraction of samples with read depth > 20x')
plt.xlabel('Reference genome position')
plt.title('subsp. abscessus (# samples with depth > 20x v. reference position)')
plt.yticks(np.arange(0, 1.05, .05))
plt.savefig('mab_depth_full')
```

### MAS

```python
with open('depth_plots/masdepth.data', 'rb') as filehandle:
    masdepth=pickle.load(filehandle)

masfrac=[i/548 for i in masdepth]
x=list(range(1,len(masdepth)+1))
```

```python
fig=plt.figure(figsize=(12,8))

plt.plot(x, masfrac, linewidth=0.25, color='orange')
plt.ylabel('Fraction of samples with read depth > 20x')
plt.xlabel('Reference genome position')
plt.title('subsp. massiliense (# samples with depth > 20x v. reference position)')
plt.savefig('mas_depth_full')
```

### BOL

```python
fig=plt.figure(figsize=(12,8))

with open('depth_plots/boldepth.data', 'rb') as filehandle:
    boldepth=pickle.load(filehandle)

bolfrac=[i/98 for i in boldepth]
x=list(range(1,len(boldepth)+1))
```

```python
fig=plt.figure(figsize=(12,8))

plt.plot(x, bolfrac, linewidth=0.25, color='green')
plt.ylabel('Fraction of samples with read depth > 20x')
plt.xlabel('Reference genome position')
plt.title('subsp. bolletii (# samples with depth > 20x v. reference position)')
plt.savefig('bol_depth_full')
```
